```python




#data processing 

```

```python
dff = pd.concat([df,pd.get_dummies(df['Month'], prefix='Month')], axis=1).drop(['Month'],axis=1)
dff = pd.concat([dff,pd.get_dummies(dff['VisitorType'], prefix='VisitorType')], axis=1).drop(['VisitorType'],axis=1)
print(dff.info())
```

```python
# data split
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=dff['Revenue']
```

```python
x=dff.drop('Revenue',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
# training the datase




```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
models=[]
cs=[0.001,0.01,0.1,1.0,10.0,100.0,1000.0,10000.0]
for i in range(len(cs)):
    model=LogisticRegression(C=cs[i])
    model.fit(x_train,y_train)
    models.append(model)
```

```python
models
```

```python
#logistic regression analyis
```

```python
x=[ model.score(x_test,y_test)  for model in models]
```

```python
x           # we can conclude inverse regression strength of 1000 so we continue with that ....
```

```python
from sklearn.metrics import confusion_matrix,classification_report 
from sklearn import metrics
```

```python
model_6=models[6]
predict_values=model_6.predict(x_test)
print('logistic regression')
print("accuracy for the model bulit with c 1000        ====",metrics.accuracy_score(y_test,predict_values))
print('F1 Score                   : ', metrics.f1_score(y_test,predict_values))
print('Precision                  : ', metrics.precision_score(y_test,predict_values))
print('Recall                     : ', metrics.recall_score(y_test,predict_values))
print('Confusion Matrix:     \n '       , confusion_matrix(y_test,predict_values))





print('classification report \n\n',metrics.classification_report(y_test,predict_values))
```
